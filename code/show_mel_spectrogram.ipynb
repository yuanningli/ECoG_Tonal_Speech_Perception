```python
import librosa
import numpy as np
from matplotlib import pyplot as plt
import os

sub = 'sub-01'
session = '01'
period = [0, 10]
mel_dB = np.load(os.getcwd().rstrip('/code')+ f'/derivatives/mel_dB/{sub}/{sub}_task-audio_run-{session}_mel_dB.npy')
plt.figure(figsize=(10, 4))
librosa.display.specshow(mel_dB[:,int(period[0]*100):int(period[1]*100)], sr=16000, hop_length=160, x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f dB')
plt.title('Mel-frequency spectrogram (dB)')
plt.savefig(f"{sub}_session-{session}_{period[0]}-{period[1]}_mel_spectrogram.svg", format="svg", dpi=300, bbox_inches="tight")
plt.tight_layout()
plt.show()
```
